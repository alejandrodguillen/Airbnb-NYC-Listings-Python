```python
#Importing Libraries

import pandas as pd
import numpy as np
import os

import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
```

```python
#Importing data

path = r'/Users/alejandroguillen/Documents/Airbnb'
```

```python
df = pd.read_csv(os.path.join(path,'Data', 'Prepared','listings_cleaned.csv'), index_col = False)
```

```python
df.head()
```

```python
df.drop('license', axis=1)
```

```python
# Get the number of rows
n_rows = len(df)
```

```python
# Get the number of rows where price is above 950
n_rows_above_950 = len(df[df['price'] > 950])
```

```python
# Calculate the percentage
percentage = (n_rows_above_950 / n_rows) * 100
```

```python
print('The number of rows is', n_rows)
print('The number of rows where price is above 950 is', n_rows_above_950)
print('The percentage of rows where price is above 950 is', percentage, '%')
```

```python
# Drop the rows where price is above 950
thresh = int(len(df) * (100 - percentage) / 100)
df = df.drop(df[df['price'] > 950].index[:thresh], axis=0)
```

```python
print(df)
```

```python
# Get the number of rows
n_rows_2 = len(df)
```

```python
# Get the number of rows where price is above 950
n_rows_above_950_2 = len(df[df['price'] > 950])
```

```python
# Calculate the percentage
percentage_2 = (n_rows_above_950_2 / n_rows_2) * 100
```

```python
print('The number of rows is', n_rows_2)
print('The number of rows where price is above 950 is', n_rows_above_950_2)
print('The percentage of rows where price is above 950 is', percentage_2, '%')
```
