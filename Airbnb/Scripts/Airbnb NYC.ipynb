```python
df_listings2 = pd.read_csv(os.path.join(path,'listings2.csv'))
```

#### Note: the original data file was imported to Excel in order to perform a faster check and cleaning in the host_name column.  20 entries were deleted because host name was in blank or not clear.

```python
# Drop the 'license' column
df_listings2 = df_listings2.drop(columns=['license'])
```

```python
# Check result 

df_listings2.head()
```

```python
# Rename the 'id' column to 'listing_id'
df_listings2 = df_listings2.rename(columns={'id': 'listing_id'})
```

```python
# Check result

df_listings2.head()
```

```python
df_listings2.info()
```

```python
# Checking for missing values

missing_values = df_listings2.isnull().sum()

# Displaying the count of missing values for each column

print(missing_values)
```

#### Note: last_review and reviews_per_month have the same amount of missing values. They are related to the fact that these hosts have zero reviews (both in general and also in the past 12 months). A lot of of them also have no availability. Will leave it like that for now.

```python
# Checking data types

df_listings2.dtypes
```

```python
# Fill blank cells in 'last_review' and 'reviews_per_month' columns with 0
df_listings2['last_review'].fillna(0, inplace=True)
df_listings2['reviews_per_month'].fillna(0, inplace=True)

# Print the DataFrame to verify the changes
print(df_listings2)
```

```python
# Checking results

missing_values = df_listings2.isnull().sum()

# Displaying the count of missing values for each column

print(missing_values)
```

#### Note: blank cells of 'last_review' and 'reviews_per_month' were replaced with 0 values.

```python
# Convert 'listing_id' and 'reviews_per_month' columns to int64
df_listings2['listing_id'] = df_listings2['listing_id'].astype('int64')
df_listings2['reviews_per_month'] = df_listings2['reviews_per_month'].astype('int64')

# Print the DataFrame to verify the data type changes
print(df_listings2.dtypes)
```

```python
#Exporting data

df_listings2.to_csv('/Users/alejandroguillen/Documents/Airbnb_listings.csv', index=False)
```
